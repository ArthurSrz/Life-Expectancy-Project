### Subjective health : do people feel better over time ?

To address this issue, I will plot a line graph with the variable `perceived_health` and see if this values increases or decreases over time. In order to do this, let's interact with the data in the following way : 
* clean the perceived_health_data dataframe : delete and rename columns, rename "United States" in "United States of America"  
* select the best type of join
* perform a left join, the left dataframe being the first "df"
* join the two dataframes on the column "Country" and "Year"
* as there will be many missing data, try to complete the dataset 

```python
df_perceived_health = pd.read_csv('perceived_health_data.csv')
df_perceived_health = df_perceived_health[df_perceived_health['Type of indicator'] == 'Average']
df_perceived_health = df_perceived_health.drop(columns = ['LOCATION','TYPE_VAR','Type of indicator', 'VARIABLE','Indicator','WB','Current/Future Well-being','SEX','Sex','AGE','Age','EDUCATION','Education','Time','Flag Codes','Flags'])
df_perceived_health = df_perceived_health.rename(columns = {"TIME":"Year"})
df_perceived_health = df_perceived_health.replace({'Country': {'United States': 'United States of America'}})
```

```python
final_df = pd.merge(df,df_perceived_health, how='left',on=['Country','Year'])
final_df = final_df.rename(columns = {'Value':'perceived_health'})
final_df.to_csv('final_data.csv')
print(final_df.info())
```

As you can see above, we **only have 29 non-null values in the column "perceived health"** (about 30% of the length of the final dataframe). It means we'll have an incomplete line graph (below)

```python
sns.lineplot(x='Year', y='perceived_health', data=final_df, hue='Country',style='Country',markers = True)
plt.legend(loc=2,bbox_to_anchor = (1,1))
plt.show()
```
